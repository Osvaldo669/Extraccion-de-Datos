```python
import pandas as pds
import numpy as np
import matplotlib.pyplot as plt
import os

datos = pds.read_csv(r"Document\\covid_19_data.csv")
resultado= datos[(datos['Confirmed']>0) & (datos['Deaths']>0) & (datos['Recovered']>0)]
pais = resultado.groupby('Country/Region').sum()
final=pais.sort_values(by=['Confirmed'],ascending=False)
lista=final.drop('SNo',axis=1)
lista_2 = lista.drop([lista.index[0]])
lista_2.plot(kind='bar', xlabel='Paises')
```
