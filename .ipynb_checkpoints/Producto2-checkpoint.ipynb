```python
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_csv('Document\AirBnB_listings.csv', encoding='unicode_escape')
datos['price'] = datos[['price']].replace(['[\$]',','],'',regex=True).astype(float)
resultado=datos[datos["neighbourhood"]=="Amsterdam, North Holland, Netherlands"]

auxiliar  = resultado.groupby("neighbourhood_cleansed")

estadisticas = resultado.groupby("neighbourhood_cleansed").price.agg(['max','min','median','mean'])
rating = resultado.groupby("neighbourhood_cleansed").review_scores_value.mean()
avg_recamaras=(auxiliar['bedrooms'].sum())/(auxiliar['bedrooms'].count())

muestra=pd.concat([estadisticas,avg_recamaras,rating], axis=1)

muestra.rename(columns={'max':'Precio maximo','min':'Precio minimo', 'median':'Mediana', 'mean':'Promedio','review_scores_value':'rating'})
```

```python
mas_baratos=muestra.nsmallest(5,'mean')
mas_baratos['mean'].plot(kind='bar',xlabel='Distritos',title='Mas Baratos')
```

```python
mas_costosos=muestra.nlargest(5,'mean')
mas_costosos['mean'].plot(kind='bar',xlabel='Distritos',title='Mas Costosos')
```
