```python
import pandas as pd

datos = pd.read_csv('Document\AirBnB_listings.csv', encoding='unicode_escape')
datos[datos['price']] = datos[['price']].replace(['[\$]',','],'',regex=True).astype(float)
datos['price']
resultado=datos[datos["neighbourhood"]=="Amsterdam, North Holland, Netherlands"]
resultado.groupby("neighbourhood_cleansed").price.agg(['max','min','median'])
```
